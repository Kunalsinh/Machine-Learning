# Data Preprocessing

```python
import numpy as np
data=np.array([[1,2],[3,4],[5,6]])
print(data)
```

```python
#Mean
print(data.mean(axis=0))
```

```python
#standard deviation
print(data.std(axis=0))
```

### Mean Removal

There are two steps to perform the Mean Removal::::
* Step: 1 Find the Mean
* Step:2 Subtract the Mean from the data

```python
data=np.array([[1,2],[3,4],[5,6]])

#step:1
mean=data.mean(axis=0)
#step:2
data_mean_removed=data-mean

print(data_mean_removed)
print(data_mean_removed.mean(axis=0))
print(data_mean_removed.std(axis=0))
```

```python
#WAY: 2 MEAN REMOVAL
from sklearn import preprocessing
print(data)
data_mean_removed1=preprocessing.scale(data)
print(data_mean_removed1)
print(data_mean_removed1.mean(axis=0))
print(data_mean_removed1.std(axis=0))
```

# Classification   : Naive Byes Classifier

```python
from sklearn import model_selection
import sklearn.datasets as db
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#model Evalution

#load the dataset : iris
dataset=db.load_iris()

#Identify X:sepal length and Y: class of flower
x=np.array(dataset.data[:,0]).reshape(-1,1)
y=np.array(dataset.target).reshape(-1,1)

#Split the dataset into training & testing
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.3)

#Load the classifier
classifier=GaussianNB()

#Train the model
classifier.fit(xtrain,ytrain)

#test the classifier
ytest_predicted=classifier.predict(xtest)

print(xtest)
print(ytest)
print(ytest_predicted)
```

## Model Visualization

```python
import matplotlib.pyplot as plt

#Model Visualization
plt.scatter(xtest,ytest,label="Actual values")
plt.scatter(xtest,ytest_predicted,label="Predicted values")
plt.xlabel("Sepal Length")
plt.ylabel("Flower Class")
plt.title("Classification Visualization of Iris Flowers")
plt.legend()
plt.show()
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

#Evaluation Metrics
print("Accuracy Score::",accuracy_score(ytest,ytest_predicted))
print("\nClassification Report::")
print(classification_report(ytest,ytest_predicted))
print("\nConfusion Metix::")
print(confusion_matrix(ytest,ytest_predicted))
```

## Logistic Regression Classifier

```python
from sklearn import model_selection
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
#model Evalution

#load the dataset : cancer
dataset=load_breast_cancer()


#Identify X and y
x=np.array(dataset.data[:,0]).reshape(-1,1)
y=np.array(dataset.target).reshape(-1,1)

#Split the dataset into training & testing
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.3)

#Load the classifier
classifier=LogisticRegression()

#Train the model
classifier.fit(xtrain,ytrain)

#test the classifier
ytest_predicted=classifier.predict(xtest)

print(xtest)
print(ytest)
print(ytest_predicted)

#Visualize the results
plt.scatter(xtest,ytest)
plt.scatter(xtest,ytest_predicted)
plt.show()

#Model Evaluation
from sklearn.metrics import confusion_matrix

print("\nConfusion Metix::")
print(confusion_matrix(ytest,ytest_predicted))
```

## SVM

```python
from sklearn import model_selection
import sklearn.datasets as db
from sklearn.svm import SVC
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#load the dataset : iris
dataset=db.load_iris()

#Identify X:sepal length and Y: class of flower
x=np.array(dataset.data[:,0]).reshape(-1,1)
y=np.array(dataset.target).reshape(-1,1)

#Split the dataset into training & testing
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.3)

#Load the classifier
classifier = SVC(kernel='linear')  
#Train the model
classifier.fit(xtrain,ytrain)

#test the classifier
ytest_predicted=classifier.predict(xtest)

print(xtest)
print(ytest)
print(ytest_predicted)

plt.scatter(xtest,ytest)
plt.scatter(xtest,ytest_predicted)
plt.show()
```
