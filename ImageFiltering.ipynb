# Image Filtering

### Reading and Displaying Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("test.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Applying kernal to an image (2d Filtering)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("test.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#k1=np.array([[0,0,0],[0,1,0],[0,0,0]])
k1=np.array([[-1,-1,-1],[-1,12,-1],[-1,-1,-1]])

#filter2D(image,depth,kernal)
filtered_img=cv2.filter2D(img,-1,k1)

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(filtered_img),plt.axis('off')
```

### Blur an Image using built-in function

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("test.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#blur(image,kernal_size(x,y))
blur_img=cv2.blur(img,(7,7))

plt.subplot(1,2,1),plt.imshow(img),plt.title("Original Image"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(blur_img),plt.title("Blurred Image"),plt.axis('off')
```

### Different types of Blurring Operations

```python
img=cv2.imread("img1.jpeg")

img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#box blur
blurred_img1=cv2.blur(img,(10,10))

#gaussian blur
blurred_img2=cv2.GaussianBlur(img,(5,5),0)

#median blur
blurred_img3=cv2.medianBlur(img,3)

plt.figure(figsize=(15,6))
#subplot(row,col,sequence)
plt.subplot(2,2,1),plt.imshow(img),plt.title("Original Image"),plt.title("Original Image"),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(blurred_img1),plt.title("Box Blur"),plt.title("Original Image"),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(blurred_img2),plt.title("Gaussian Blur"),plt.title("Original Image"),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(blurred_img3),plt.title("Median Blur"),plt.title("Original Image"),plt.axis('off')
```

## Edge Detection

```python
img=cv2.imread("test.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15,6))

#sobel(image,ddepth,x,y,ksize)
#ddpeth: output image size/rsolution
#x=1:horizontally
#y=1: vertically

sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,5)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,5)

laplacian_img=cv2.Laplacian(img,cv2.CV_64F)

#canny(image,min-range,max-range)
canny_img=cv2.Canny(img,50,150)


plt.subplot(1,4,1),plt.imshow(sobelx,cmap="gray"),plt.title("Sobel Horizontal"),plt.axis('off')
plt.subplot(1,4,2),plt.imshow(sobely,cmap="gray"),plt.title("Sobel Verical"),plt.axis('off')
plt.subplot(1,4,3),plt.imshow(laplacian_img,cmap="gray"),plt.title("Laplacian"),plt.axis('off')
plt.subplot(1,4,4),plt.imshow(canny_img,cmap="gray"),plt.title("Canny"),plt.axis('off')
```

# Erode & Dilate

```python
img=cv2.imread("images.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1=np.ones((5,5))

#erode(image,kernal,iterations=layers)
#dilate(image,kernal,iterations=layers)

img_erode=cv2.erode(img,k1,iterations=1)
img_dilate=cv2.dilate(img,k1,iterations=1)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1),plt.imshow(img),plt.title("original image")
plt.subplot(1,3,2),plt.imshow(img_erode),plt.title("Erode image")
plt.subplot(1,3,3),plt.imshow(img_dilate),plt.title("Dilate image")
```

# Contrast Effect on Color image

```python
#Reading an image
img=cv2.imread("img1.jpeg")

#convert the image in RGB format
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#convert the image into YUV format
img_YUV=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)

#split the image
img_YUV[:,:,0]=cv2.equalizeHist(img_YUV[:,:,0])


#Convert YUV image into Color
final_img=cv2.cvtColor(img_YUV,cv2.COLOR_YUV2BGR)

plt.subplot(1,2,1),plt.imshow(img),plt.title("original image")
plt.subplot(1,2,2),plt.imshow(final_img),plt.title("Contrast image")

```

## Motion Blur

```python
#Motion Blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("tree.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

size=100

#Horizontal Motion Blur
k1 = np.zeros((size, size))        #creating a kernal with zero value of given size
k1[int(size/2),:]=np.ones(size)    #find the middle row of the kernal and assign a new array of ones
k1 = k1/ size                      #now devide that perticular value according to the size of the image


#Vericall Motion Blur
k2 = np.zeros((size, size))
k2[:,int(size/2)]=np.ones(size)
k2 = k2/ size

h_img=cv2.filter2D(img,-1,k1)
v_img=cv2.filter2D(img,-1,k2)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1),plt.imshow(img),plt.title("original image")
plt.subplot(1,3,2),plt.imshow(h_img),plt.title("Horizontal Motion Blur")
plt.subplot(1,3,3),plt.imshow(v_img),plt.title("Vertical Motion Blur")
```

## Embossing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.cvtColor(cv2.imread('img1.jpeg'),cv2.COLOR_BGR2RGB)

k1 = np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
k2 = np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
k3 = np.array([[1,0,0],[0,0,0],[0,0,-1]])
k4 = np.array([[0,0,1],[0,0,0],[-1,0,0]])

img1 = cv2.filter2D(img,-1,k1) + 128
img2 = cv2.filter2D(img,-1,k2) + 128
img3 = cv2.filter2D(img,-1,k3) + 128
img4 = cv2.filter2D(img,-1,k4) + 128

plt.figure(figsize=(15,6))
plt.subplot(2,3,1),plt.imshow(img)
plt.subplot(2,3,2),plt.imshow(img1)
plt.subplot(2,3,3),plt.imshow(img2)
plt.subplot(2,3,4),plt.imshow(img3)
plt.subplot(2,3,5),plt.imshow(img4)
```
