## Clustering

### KMeans Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Load the  CSV file
data = pd.read_csv("Book1.csv")
x=data.iloc[:,0]
y=data.iloc[:,1]

#plot the dataset to identify the clusters 
plt.figure()
plt.scatter(x, y, marker='o',facecolors='none', edgecolors='k', s=30)
plt.title('Input data')
plt.xticks(())
plt.yticks(())

#Perform the KMeans CLustering 
#Load the Model
kmeans=KMeans(init='k-means++',n_clusters=4)

#Train the Model
kmeans.fit(data)

#Use the Model to identify the clusters 
results=kmeans.predict(data)
print("Clustering Results")
print(results)
print("="*50)

#Find the centroids 
centroids = kmeans.cluster_centers_
print("Centroids")
print(centroids)

#plot the clusters along with centroids
plt.figure()
plt.scatter(x, y,marker='o',facecolors='none',edgecolors='k',s=30)
plt.scatter(centroids[:,0],centroids[:,1],marker='o',s=200,color='r')
plt.title('Centroids on boundaries obtained using KMeans')
```

```python
#KMeans Algorithm : Unsupervised Learning (Clustering)
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.clusters import KMeans

#Step : 1 Load the data 
plt.figure()
data=pd.read_csv("Book1.csv")
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(x,y)

#Step: 2 Load the Model
model=KMeans(init="k-means++",n_clusters=4)

#Step : 3 Train the model
model.fit(data)

#Use the model to display the clusters
results=model.predict(data)
print(results)

#find the centroids (center point)
centroids = model.cluster_centers_
print(centroids)
```

## Image Compression using Clustering

```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

image=io.imread('hqdefault.jpg')
image_r=(image/255.0).reshape(-1,3)
k_colors=KMeans(n_clusters=2).fit(image_r)
img128=k_colors.cluster_centers_[k_colors.labels_]
img128=np.reshape(img128,(image.shape))
img.imsave("img.jpg",img128)
```
